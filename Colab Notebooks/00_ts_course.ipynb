```python
tf.random.set_seed(1234)

@tf.function
def f():
  a = tf.random.uniform([1])
  b = tf.random.uniform([1])
  return a, b

@tf.function
def g():
  a = tf.random.uniform([1])
  b = tf.random.uniform([1])
  return a, b

print(f())  # prints '(A1, A2)'
print(g())  # prints '(A1, A2)'
```

```python
@tf.function
def foo():
  a = tf.random.uniform([1], seed=1)
  b = tf.random.uniform([1], seed=1)
  return a, b
print(foo())  # prints '(A1, A1)'
print(foo())  # prints '(A2, A2)'

@tf.function
def bar():
  a = tf.random.uniform([1])
  b = tf.random.uniform([1])
  return a, b
print(bar())  # prints '(A1, A2)'
print(bar())  # prints '(A3, A4)'
```

### Aggregating tensors

Aggegating tensors = condensing them from mult values down to a smaller amount of values

```python
# get the absolute values
F = tf.constant([-7, -10])
tf.abs(F)
```

```python
G = tf.constant(np.random.randint(0, 100, size=50))
G
```

```python
print(tf.reduce_min(G))
print(tf.reduce_max(G))
print(tf.reduce_mean(G))
print(tf.reduce_sum(G))
```

```python
# find the variance of our tensor
import tensorflow_probability as tfp
tfp.stats.variance(G)
```

```python
# standart deviation
tf.math.reduce_std(tf.cast(G, dtype=tf.float32)) #it works just to float tensors
```

# find the positional minimum n maximum of a tensor (argmin n argmax)

```python
tf.random.set_seed(42)
I = tf.random.uniform(shape=[50])
I
```

```python
# max
tf.argmax(I) # same as np.argmax() // return position
```

```python
# index on our largest value position
I[tf.argmax(I)]
```

```python
tf.reduce_max(I)
```

```python
I[tf.argmax(I)] == tf.reduce_max(I)
```

```python
tf.argmin(I)
```

```python
I[tf.argmin(I)]
```

```python
I[tf.argmin(I)] == tf.reduce_min(I)
```
